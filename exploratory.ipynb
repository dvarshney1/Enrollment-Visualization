```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset.csv", sep = ",")
```

```python
df.head()
```

```python
df = df.sort_values(by = "Fall")
```

```python
df.head()
```

```python
df = df.loc[(df["Fall"] == 1980) | (df["Fall"] == 2018)]
```

```python
df.shape
```

```python
df = df.reset_index()
```

```python
list(df)
```

```python
columns_to_drop = ["index", 
                   "Term Code", 
                   "Department", 
                   "Degree", 
                   "Major Code", 
                   "Concentration Code", 
                   "Concentration Name",
                   'Unknown',
                   'Caucasian',
                   'Asian American',
                   'African American',
                   'Hispanic',
                   'Native American',
                   'Hawaiian/Pacific Isl',
                   'Multiracial',
                   'Unknown.1',
                   'All African American',
                   'All Native American',
                   'All Hawaiian/ Pac Isl',
                   'All Asian',
                   'Academic Program Code',
                   'Legacy Advising Code',
                   'Legacy Department Code',
                   "Legacy Major Name", 
                   "Illinois", 
                   "Non-Illinois", 
                   "International"
                  ]

df = df.drop(columns_to_drop, axis = 1)
```

```python
df = df.loc[df["Major Name"] != df["Major Name"].str.upper()]
```

```python
df = df.dropna()
```

```python
df['Total'] = df['Total'].astype(str).str.replace(",","").astype(int)
df['Male'] = df['Male'].astype(str).str.replace(",","").str.replace(".","").astype(int)
df['Female'] = df['Female'].astype(str).str.replace(",","").str.replace(".","").astype(int)
```

```python
df.dtypes
```

```python
df_1980 = df.loc[df["Fall"] == 1980]
df_2018 = df.loc[df["Fall"] == 2018]
```

```python
df_1980 = df_1980.groupby(["Major Name", "College"])["Total", "Male", "Female"].sum()
df_1980["Fall"] = 1980
df_1980 = df_1980.reset_index()

df_2018 = df_2018.groupby(["Major Name", "College"])["Total", "Male", "Female"].sum()
df_2018["Fall"] = 2018
df_2018 = df_2018.reset_index()
```

```python
df_1980.shape
```

```python
df_2018.shape
```

```python
df = pd.concat([df_1980, df_2018])
```

```python
df = df.reset_index()
```

```python
df = df.drop(["index"], axis = 1)
```

```python
counts = df["Major Name"].value_counts()
```

```python
df = df[df["Major Name"].isin(counts[counts > 1].index)]
```
